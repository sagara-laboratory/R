```r
DF <- read.table( "sample-height.csv", 
                  sep = ",",                
                  header = TRUE,            
                  stringsAsFactors = FALSE) 
```

```r
str(DF)
head(DF)
hist(DF$height)
```

```r
hist(DF$height, col="steelblue", breaks=50,
     main="")
```

```r
library(ggplot2) 
ggplot(DF)+                         
  geom_histogram( aes(height),      
                  fill="steelblue",  
                  alpha=0.8,        
                  binwidth=1 )
```

```r
ggplot(DF)+                          
  geom_density( aes(height) ) +         
  theme(axis.text=element_text(size=12),
       axis.title=element_text(size=14,face="bold"))
```

```r
ggplot(DF)+                          
  geom_histogram( aes(x   =height,     
                      fill=gender),    
                  position="identity", 
                  alpha=0.5,           
                  binwidth=1 )         
```

```r
ggplot(DF)+                          
  geom_histogram( aes(x   =height,     
                      fill=gender),    
                  position="identity", 
                  alpha=0.5,           
                  binwidth=1 )+
  scale_fill_manual( values=c("darkgreen", "orange") )+
  theme(axis.text =element_text(size=12, face="bold"), 
        axis.title=element_text(size=14) )      
```

```r
ggplot(DF)+                          
  geom_density( aes(x    =height,      
                    color=gender) )+   
  scale_color_manual( values=c("darkgreen", "orange") )
```

```r
ggplot(DF)+                          
  geom_density( aes(x    =height,      
                    color=gender,      
                    fill =gender),     
                    alpha=0.3  )+      
  scale_color_manual( values=c("darkgreen", "orange") )+
  scale_fill_manual(  values=c("darkgreen", "orange") )+
  theme(axis.text =element_text(size=12, face="bold"), 
        axis.title=element_text(size=14) )
```

```r
boxplot(DF$height ~ DF$gender,       
                     col="orange")
```

```r
boxplot(DF$height ~ DF$gender,       
                     col="orange",     
                     horizontal=T,     
        main="高校生の身長の分布") 
```

```r
ggplot(DF)+                          
  geom_boxplot( aes(y    =height,      
                    x    =gender,      
                    fill =gender),     
                    alpha=0.7 ) 
```

```r
ggplot(DF)+                          
  geom_violin(  aes(y    =height,      
                    x    =gender,      
                    fill =gender),     
                    alpha=0.5 )        
```

```r
ggplot(DF)+                          
  geom_violin(  aes(y    =height,      
                    x    =gender,      
                    fill =gender),     
                    alpha=0.5 )+       
  geom_boxplot( aes(y    =height,      
                    x    =gender),     
                    fill ="grey",      
                    width=.2,         
                    alpha=0.7)+   
  scale_color_manual( values=c("darkgreen", "orange") )+
  scale_fill_manual(  values=c("darkgreen", "orange") )
```
